```python
def polar_curve(fi):
    if(fi<pi/3): return sin_circle(fi, modulation=0.25*fi, period=2*np.pi/10)
    if(fi<3*pi/2): return sin_circle(fi*2, modulation=0.05, period=2*np.pi/10)
    return sin_circle(fi, modulation=0.07, period=2*np.pi/10)
evolve_agat(make_xy_list(polar_curve, 1000),min_area=0.1,layer_width=0.02)
```

```python
np.random.seed(12345)
modulation=0.3
period=2*pi*1.5
def polar_curve(fi):
    r_0 = 1/(1 + modulation) #average radius
    a = r_0*modulation #amplitude of sin
    freq = (2*np.pi)/period
    return r_0 * (1 + a*np.sin((freq*fi)**2))

evolve_agat(make_xy_list(polar_curve, 1000),min_area=0.01,layer_width=0.02)
```
